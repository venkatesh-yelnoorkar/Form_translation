```python
import numpy as np
import pandas as pd
```

```python
from google.cloud import translate
```

```python
#initialize translation client
translate_client = translate.Client()
```

```python
#input
df = pd.read_excel("test.xlsx", sheet_name = "Sheet1", header=None)
```

```python
#get dimensions of the csv
size = df.shape
```

```python

#replace mising values by ' '(space)
df.fillna(value = ' ', inplace = True)
```

```python
df
```

```python
#iterrate or each row and col and get and send the contents of each cell to the translator
for i in range(0, size[1]):
    for j in range(0, size[0]):
        temp = translate_client.translate(df.iloc[j, i], target_language = 'en')
        df.iloc[j, i] = temp["translatedText"]
        
```

```python
df
```

```python
#output
df.to_csv('output', index = False, header = False)
```
